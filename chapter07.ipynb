## 正文
现我们来举例求解$Ax=0$。 
下面举例;
$A=\begin{bmatrix}1&2&2&2\\2&4&6&8\\3&6&8&10\end{bmatrix}$  
观察它的行与列，可以发现列二是列一的两倍，他们是一个方向的向量，列一与列二相关；
行一+行二=行三，说明行三与行一，行二相关；  
这些都会在消元时表现出来！  

有一点，对于$Ax=0$,我们在对A进行消元时，不会影响到方程组的解,因为右侧为0向量。  

$A=\begin{bmatrix}1&2&2&2\\0&0&2&4\\0&0&2&4\end{bmatrix}$-》
$A=\begin{bmatrix}1&2&2&2\\0&0&2&4\\0&0&0&0\end{bmatrix}$  

这里我们看到，这个矩阵消元后，主元的数量为**2**,分别是$a_{11},a_{23}$ 这个数字我们称为矩阵的秩(rank of matrix)。
rank of A  
= num of pivots  
=2  

现在我们将$Ax=0$变为了$Ux=0$,接下来我们就进行回代，看如何求解  
现在我们知道有2个主元，主元所在的列是 主列(pviot column)，非主列称为 自由列(free column).  
这里自由是什么意思？这里自由列表示可以任意分配数值给属于该列的未知数。  
就是说对于$A=\begin{bmatrix}1&2&2&2\\0&0&2&4\\0&0&0&0\end{bmatrix}$  ,
列1与列3是主列，列二与列四是自由列。所以对于这个方程组而言$x_2$与$x_4$可以取任意数，然后再求解$x_1$与$x_3$。  
将矩阵$U$，转换为方程组:  
$\begin{cases}x_1+2x_2+2x_3+2x_4=0\\2x_3+4x_4=0\end{cases}$  
这里$x_1$与$x_3$我们可以通过回代求得，这里新的知识点就是**自由变量(free variable)**，这里我们选择$x_2=1$和$x_4=0$：
$x=\begin{bmatrix}x_1\\1\\x_3\\0\end{bmatrix}$->$x=\begin{bmatrix}-2\\1\\-1\\0\end{bmatrix}$  
这是零空间的一个向量，为了得到更多这个方向上的向量，我们可以在它前面加个倍数$x=c\begin{bmatrix}-2\\1\\-1\\0\end{bmatrix}$  
我们再将$x_2=0$，$x_4=1$得到：$x=\begin{bmatrix}2\\0\\-2\\1\end{bmatrix}$ 这个解表示：
2个列一减2一个列三加1一个列四确实为0，这就是零空间的另外一个向量  
通过这两个向量就能求出整个零空间的解了，这两个解我们称为特解（Special solutions），怎么个特殊法呢?
特殊的地方就是给自由变量分配的是0和1.  
通过这两个特解，我们得到零空间或者说$Ax=0$的解，就是这两个特解的线性组合 $x=c\begin{bmatrix}-2\\1\\-1\\0\end{bmatrix}+d\begin{bmatrix}2\\0\\-2\\1\end{bmatrix}$

教授得到的结论就是：零空间所包含的正好是特解的线性组合。那么有多少个特解呢？ 每个自由变量对应一个特解！！  
对于mxn矩阵，秩为r，那么就有n-r个自由变量。

对于求解ax=0，我们给出了一整套算法：
首先是消元，碰到无法取得主元的列，不管它，继续消元，之后得到关键的主元数r，剩下n-r个自由变量，然后另它为1,0.

下面我们说简化行阶梯形式。$R$=reduced row echelon form   
$U=\begin{bmatrix}1&2&2&2\\0&0&2&4\\0&0&0&0\end{bmatrix}$，行三全为0是因为行但是行一行二的线性组合(combination),
消元发现了这点。那么怎么继续简化呢，我可以向上简化，因此简化行阶梯形式中，主元上下全是0.现在我们来进行简化    
$U=\begin{bmatrix}1&2&2&2\\0&0&2&4\\0&0&0&0\end{bmatrix}$->$\begin{bmatrix}1&2&0&2\\0&0&2&4\\0&0&0&0\end{bmatrix}$->$U=\begin{bmatrix}1&2&0&2\\0&0&1&2\\0&0&0&0\end{bmatrix}=R=rref(A)$  
在Matlab里用命令rref命令瞬间就能完成这些动作，rref(A)是Reduced Row Echelon Form of A的意思A的简化行阶梯形式。   
它以最简形式包含了所有信息，包含了什么信息？  
- 我们可以看出主行，这里是行一，行二；  
- 主列，这里是列一，列三；  
- 实际上它还包含了一个单位矩阵，在主行与主列的交汇处。  
- 最后还有一行0，原行是其他行的线性组合；  
- 得到了自由列；  

这里特解也就很容易看出了；
我们从$Ax=0$中间转换为$Ux=0$，以及最终的$RX=0$,解均相同。因为在消元过程中没有改变解。  

对于$R$转换来的方程组$\begin{cases}x_1+2x_2+2x_4=0\\x_3+2x_4=0\end{cases}$  ,
我们可以把主列与自由列分开处理， 
主列为$\begin{bmatrix}1&0\\0&1\end{bmatrix}$记为$I$  
自由列为$\begin{bmatrix}2&-2\\0&2\end{bmatrix}$记为$F$  
0那一行可以忽略。  

这样看$R$就为$\begin{bmatrix}I&R\\0&0\end{bmatrix}$，我们的零空间$N$也就是方程组$Rx=0$的解，即$RN=0$,$N=\begin{bmatrix}-F\\I\end{bmatrix}$ 这就是特解求的矩阵，matlab用null命令来求出，null（A）得到一个零空间。  

matlab是怎么求出的呢？  
先求出R，然后找出主变量与自由变量， 将1和0分配到自由变量中，复制出主变量， 它会使用回代，我们这里的回代非常简单.  
$\begin{bmatrix}I&F\end{bmatrix}$$\begin{bmatrix}x_{pivot}\\x_{free}\end{bmatrix}=0$，这里得到一个等式：  
$x_{pivot}=-Fx_{free}$，这是回代最终得到的结果，不断简化得到的最简形式， 如果给$x_{free}$分配一个单位矩阵，那么主变量就变为了$-F$.  

我们再举个例子，用这个算法再算遍,这次我们取$A$的转置$\begin{bmatrix}1&2&3\\2&4&6\\2&6&8\\2&8&10\end{bmatrix}$  
有多少个pivot variables？
做它的rref：  
$\begin{bmatrix}1&2&3\\2&4&6\\2&6&8\\2&8&10\end{bmatrix}$->$\begin{bmatrix}1&2&3\\0&0&0\\0&2&2\\0&4&4\end{bmatrix}$->$\begin{bmatrix}1&2&3\\0&2&2\\0&0&0\\0&0&0\end{bmatrix}=U$  
r=2 ! 这里有个结论：矩阵主列的个数与其转置相同。free variables num = 3-2 =1  
另自由变量为1得到
$x=\begin{bmatrix}-1\\-1\\1\end{bmatrix}$ ，整个零空间就是一条直线，$x=c\begin{bmatrix}-1\\-1\\1\end{bmatrix}$ 
零空间的基(basic for the null space)就是向量$\begin{bmatrix}-1\\-1\\1\end{bmatrix}$。

求$R$：
$\begin{bmatrix}1&2&3\\0&2&2\\0&0&0\\0&0&0\end{bmatrix}=U$->$\begin{bmatrix}1&0&1\\0&1&1\\0&0&0\\0&0&0\end{bmatrix}=R=\begin{bmatrix}I\&F\\0&0\end{bmatrix}$
从$R$可以看出$F=\begin{bmatrix}1\\1\end{bmatrix}$,再看$x=c\begin{bmatrix}-1\\-1\\1\end{bmatrix}=c\begin{bmatrix}-F\\I\end{bmatrix}=N$
