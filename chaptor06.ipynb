## 正文
首先回忆一下向量空间是什么？
这里用一句话总结：
一组向量，他们的所有**加法**、**乘法**，也就是**线性组合**仍然在这组向量中，那么这组向量就构成了一个向量空间！
Vector Space requirements:
All combinations are in the space.

下面举几个例子：一些是向量空间一些不是！
一个3维空间$R^3$，很显然它是一个向量空间，满足向量空间的条件。
下面看它的子空间，他是向量空间里的(inside)的向量空间。
- 穿过原点的平面是$R^3$的子空间(Plane through $\begin{bmatrix}0\\0\\0\end{bmatrix}$ is SubSpace of $R^3$，记为$P$
- 穿过远点的直线是 $R^3$的子空间,记为$L$

现在有个问题：
我们取两个子空间P,L的并集 P∪L=all vecotrs in both P or L ，它还是子空间吗？
不是。

那么交集呢？ P∩L=all vecotrs in both P and L
....

Subspace S and T ,
intersection of S∩T is a SubSpace!!
显然，取S,T交集的两个向量V,W.V+W 属于S,也属于T ，所以它还是在S与T的交集中
对于乘法同理，所以S∩T是一个子空间。

## 列空间
这有一个矩阵$A$=$\begin{bmatrix}1&1&2\\2&1&3\\3&1&4\\4&1&5\end{bmatrix}$，它的列向量属于一个四维空间$R^4$,
这个矩阵$A$的列空间是$R^4$的子空间，记做$C(A)$。

这个子空间包含了什么？
all linear combinations of columns.

那由这3个向量的线性组合构成的空间会是整个$R^4$吗?显然不是，3个四维向量不可能得带一个四维空间
那么它就是一个较小的空间，

Does $Ax=b$ have a solution for every b.? No
因为有四个方程组，却只有3个未知数，我们把方程组写出来：

$Ax$=$\begin{bmatrix}1&1&2\\2&1&3\\3&1&4\\4&1&5\end{bmatrix}$$\begin{bmatrix}x_1\\x_2\\x_3\end{bmatrix}$=
$begin{bmatrix}b_1\\b_2\\b_3\\b_4\end{bmatrix}$.

这个方程组并不是对所有$b$有解，论证依据就是之前我们说过的：3个向量的线性组合不是完整的$R^4$，所有还有大量的向量不是这3个向量的线性组合，这些$b$时当前方程组无解。

**问题：什么b让方程组有解？？！**
b=$\begin{bmatrix}0\\0\\0\\0\end{bmatrix}$，b=$\begin{bmatrix}1\\2\\3\\4\end{bmatrix}$,b=$\begin{bmatrix}1\\1\\1\\1\end{bmatrix}$.
I can solve  $Ax=b$ exacly when b is in $C(A)$。
**当$b$在矩阵$A$的列空间中时，方程组有解。**

可以看，我们举的例子里，$col_3=col_1+col_2$，也就是说列3与列1，列2线性相关，$col_3$没有起到贡献，它就不是**主列**,$col_1$,$col_2$是主列。

## 零空间
还是用上面的例子$A$=$\begin{bmatrix}1&1&2\\2&1&3\\3&1&4\\4&1&5\end{bmatrix}$。
零空间是特殊的子空间，它包含了上面？
Null Space of A=all solutions x=$\begin{bmatrix}x_1\\x_2\\x_3\end{bmatrix}$ to $Ax=0$.
是的，我们现在对$x$兴趣了！
$Ax$=$\begin{bmatrix}1&1&2\\2&1&3\\3&1&4\\4&1&5\end{bmatrix}$$\begin{bmatrix}x_1\\x_2\\x_3\end{bmatrix}$=
$\begin{bmatrix}0\\0\\0\\0\end{bmatrix}$.
x是$R^3$下的一条直线。
$Av=0$,$Aw=0$ then $A(v+w)=0$ ，同理数乘也一样。
说明x构成一个向量空间。

$Ax$=$\begin{bmatrix}1&1&2\\2&1&3\\3&1&4\\4&1&5\end{bmatrix}$$\begin{bmatrix}x_1\\x_2\\x_3\end{bmatrix}$=
$\begin{bmatrix}1\\2\\3\\4\end{bmatrix}$.
现在所有解构成向量空间吗？No！
首先没有零向量，就无法构成向量空间。那么解构成的是什么？
在本例中它是一个不穿过原点的平面。

-----
今天我们了解到了两种构筑子空间的方法。一个通过几个向量的线性组合，也可以通过一个方程组，让x满足特定条件来得到子空间！
