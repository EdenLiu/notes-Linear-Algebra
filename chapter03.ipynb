## 概述：

矩阵乘法(4种方法)，矩阵A的逆，$A^T$和Gauss-Jodan 找$A$的逆$A^{-1}$
-----

首先我们来举一个例子：$AB=C$
我们来看一个矩阵$C$中的一个元素 $C_{34}$是这么求得的：
$C_{34}$ = （$row_3$ of $A$）* ($column_4$ of $B$) 实际公式就是
$C_{34}$=$a_{31}$$b_{14}$+$a_{32}$$b_{24}$+... = $\sum_{k=1}^{n}$$a_{3k}b_{k4}$

其中$a_{ij}$表示$A$矩阵中第i行第j列的元素，$b_{ij}$表示$B$矩阵中第i行第j列的元素

---
矩阵在什么情况下才能相乘？
对于$AB$，$A$的总列数与$B$的总行数相等，用公式表示就是：

$A_{mxn} B_{nxp}=C_{mxp}$

这就是乘法的一般性法则

矩阵乘法的整列考虑：
$\underbrace{\begin{bmatrix} & & \\ & & \end{bmatrix}_A}$*
$\underbrace{\begin{bmatrix} & \\ & \\ & \end{bmatrix}_B}$=
$\underbrace{\begin{bmatrix} & \\ & \end{bmatrix}_C}$
$A$乘以$B$的第一列，得到$C$的第一列，
$B$可以考虑为P个单独的列向量。
$C$中的列是矩阵A的线性组合，在C中我们看到的就是矩阵$A$各列的线性组合

----
整行考虑：
$A$的第一行乘以$B$，得到$C$的第一行，
$A$可以考虑为$m$个单独的行向量。
$C$中的各行是矩阵B的线性组合，在$C$中我们看到的就是矩阵B的各行的线性组合

----
列乘以行
column of $A$ x row of $B$ 可以看成是一个 mx1的矩阵乘以 1xp的矩阵得到一个mxp的矩阵
$AB$=sum of (column of $A$ x row of $B$)例如：

$\begin{bmatrix}2&5\\3&6\\4&7\end{bmatrix}$*
$\begin{bmatrix}1&2\\0&0\end{bmatrix}$ =
$\begin{bmatrix}2\\3\\4\end{bmatrix}$*$\begin{bmatrix}1&2\end{bmatrix}$+
$\begin{bmatrix}5\\6\\7\end{bmatrix}$*$\begin{bmatrix}0&0\end{bmatrix}$
=$\begin{bmatrix}2&4\\3&6\\4&8\end{bmatrix}$

----
将矩阵切成块相乘：
当然，切的时候需要相互匹配，符合能够乘法的要求，举一个例子：
矩阵$A$,$B$都是n*x*n矩阵，我们均分成4块$A_1$~$A_4$,$B_1$~$B_4$
$\begin{bmatrix}A_1&A_2\\A_3&A_4\end{bmatrix}$ $\begin{bmatrix}B_1&B_2\\B_3&B_4\end{bmatrix}$ =
$\begin{bmatrix}A_1*B_1+A_2*B_3&...\\...&...\end{bmatrix}$
，这样可以简化思维

## Inverses逆(Square Matrix)

----
我们这里首先考虑方矩阵，不是所有矩阵都可逆，逆是否存在是个非常重要的问题
我们假设$A$可逆:
$A^{-1}A=I=AA^{-1}$
假如$A^{-1}$存在，我们就说$A$是可逆的（invertible）,非奇异的（nonsingular）

下面讨论什么情况下不可逆，举个例子，举一个2X2的不可逆的矩阵

比如$\begin{bmatrix}1&3\\2&6\end{bmatrix}$

用整列乘的思想来想象下，假如有$A^{-1}A=I$，那么意思就是$A$各列的线性组合可以构成
$\begin{bmatrix}1&0\\0&1\end{bmatrix}$，但是$A$各列是线性相关的，他们在同一条直线上，
所以他们的线性组合也一定在同一条直线上，这样就不可能得到$\begin{bmatrix}1\\0\end{bmatrix}$
$\begin{bmatrix}0\\1\end{bmatrix}$，因为他们不在同一条直线上。

**I can find a nozore vector $X$ with $AX=0$,那么$A$不可逆。**
用公式证明是：$A^{-1}AX=IX=0$,而$X$是非零向量，$IX=0$不可能成立！，所以不存在$A^{-1}A=I$,$A$不可逆。

----
现在考虑存在逆的情况：
举个例子：$\begin{bmatrix}1&3\\2&7\end{bmatrix}$,如何求得它的逆？

$\begin{bmatrix}1&3\\2&7\end{bmatrix}$
$\begin{bmatrix}a&b\\c&d\end{bmatrix}$=
$\begin{bmatrix}1&0\\0&1\end{bmatrix}$

$A$* column of $A^{-1}$= column of $I$分解后就是求方程组问题了
Gauss-Jordan(solve 2 eqns at once)
本来是这种形式的$\begin{cases}\begin{bmatrix}1&3\\2&7\end{bmatrix}\begin{bmatrix}a\\b\end{bmatrix}=\begin{bmatrix}1\\0\end{bmatrix}\\\begin{bmatrix}1&3\\2&7\end{bmatrix}\begin{bmatrix}c\\d\end{bmatrix}=\begin{bmatrix}0\\1\end{bmatrix}\end{cases}$

Jordan说两个方程一起来算吧，用这种形式：
$\left[\begin{array}{cc|cc}
1&2&1&0\\3&7&0&1\end{array}\right]$
接下来用高斯的方法，对左侧进行消元，当左侧变为单位矩阵后，右侧就是$A^{-1}$的逆，下面进行消元：
$\left[\begin{array}{cc|cc}
1&2&1&0\\3&7&0&1\end{array}\right]$->
$\left[\begin{array}{cc|cc}
1&0&1&0\\3&1&-2&1\end{array}\right]$->
$\left[\begin{array}{cc|cc}
1&07&-3\\0&1&-2&1\end{array}\right]$
右侧就是$A^{-1}$,

这就是Guass-Jordan idea:
$\left[\begin{array}{c|c}A&I\end{array}\right]$->$\left[\begin{array}{c|c}I&A^{-1}\end{array}\right]$

为什么我们能这样得到逆,我们这样想：

我们把对A的所有操作看做E，我们知道了$EA=I$,所以$E=A^{-1}$,
最后右侧的结果就是$EI=E=A^{-1}$

以上就是Guass-Jordan idea！
