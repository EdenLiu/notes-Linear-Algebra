## 正文
### 矩阵乘法$AB$的逆
$AA^{-1}=I=A^{-1}A$

首先来个简单的，求矩阵$AB$的逆：

$AB$__ = $I$

根据结合律很容易得到
$(AB)(B^{-1}A^{-1})$=$A(BB^{-1})A^{-1}$=$AIA^{-1}$=$AA^{-1}$=$I$。
同样$(B^{-1}A^{-1})(AB)$=$I$

----
### 矩阵$A$的置换矩阵$A^T$
这里有个基本公式：对于可逆矩阵$A$，$A$的逆的转置就是$A$的转置的逆。
$(A^{-1})^TA^T$=$I$，
对于单个矩阵来说，转置与逆的运算是可以交换的

-----
### $A=LU$
矩阵$L$联系这个如何中矩阵$A$得到矩阵$U$,我们来看个简单的例子：2x2矩阵$A$:$\begin{bmatrix}2&1\\8&7\end{bmatrix}$

$E_{21}A$=$U$
$\begin{bmatrix}1&1\\-4&1\end{bmatrix}$$\begin{bmatrix}2&1\\8&7\end{bmatrix}$=
$\begin{bmatrix}2&1\\0&3\end{bmatrix}$ ->

$A$=$LU$
$\begin{bmatrix}2&1\\8&7\end{bmatrix}$=$\begin{bmatrix}1&1\\4&1\end{bmatrix}$$\begin{bmatrix}2&1\\0&3\end{bmatrix}$->

$A$=$LDU$
$\begin{bmatrix}2&1\\8&7\end{bmatrix}$=$\begin{bmatrix}1&1\\4&1\end{bmatrix}$$\begin{bmatrix}2&0\\0&3\end{bmatrix}$$\begin{bmatrix}1&1/2\\0&1\end{bmatrix}$
L表示Lower，U表示Upper，下三角矩阵与上三角矩阵  
这些AMTLAB都能够生成

-----
下面进入3x3矩阵情况
$E_{32}E_{31}E_{21}A$=$U$ （no row exchange）
为了得到$A=LU$，可以转换为$A=E_{21}^{-1}E_{32}^{-1}E_{31}^{-1}U$

对于$A=LU$ 如果不存在行变换，消元乘数可以直接写入$L$中。

问题：对于一个NxN矩阵，需要多少次操作？
n=100时，实际进行多少次操作，一次操作指一次乘法加减法，因为消元时肯定是乘法后再减法。为了简单，我们加其考虑为一次操作
在消第一列时，大于是$100^2$实际是100x99次操作，消一行需要进行100次操作，一共99行，但是大约是这样，性质不变。以此类推
答案是：$100^2+99^2+...+1^2$，大约为$1/3n^3$，这里对$n^2$做微积分，可以得到结果。

右侧向量$b$需要进行$n^2$次操作。
