## 第二章 消元矩阵

这是我们今天要求解的方程组：
$\begin{cases}x&+2y&=2\\3x&+8y&+z&=12\\4y&+z&=2\end{cases}$

我们用消元法(elimination)来求它的解，所有软件程序都是用消元法来求解矩阵的，当然前提是可以使用消元法能够生效，一般都能得到解，只要系数矩阵$A$是个"好矩阵"的话 ，那么问题来了，什么样的矩阵能使消元法生效，什么样的矩阵又使得消元法失效呢？在接下来的讨论中，我们会找到答案 。反之，我们也能够通过消元法知道什么矩阵是"好矩阵"，什么矩阵是"坏矩阵"。

消完元后，为了求的解，还有一个回代(back substitution)的步骤.

了解上述概念后，下面我们将要做的事是：用矩阵语言表述消元法，毕竟对于整个课程来说：
核心概念是"矩阵变换"(matrix operations)，这里我们会遇到矩阵乘法。

上述方程是形式是:$Ax=b$,那么怎么进行消元，第一步是什么？

看方程1，我们用方程1乘以某个数，然后从方程2将其减去，目的是消去方程2中的x，
我们用矩阵$\begin{bmatrix}1&2&1\\3&8&1\\0&4&1\end{bmatrix}$来看，这样更加简洁，
左上角的"1"是消元的关键，它称为"主元"(pivot)，存在主元的行称为主元行。
很明显第一行乘以3减去(subtract)第二行，可以达到目的，运算后的矩阵变为
$\begin{bmatrix}1&2&1\\0&2&-2\\0&4&1\end{bmatrix}$，这样第一步就完成了。当然有同学会问右侧向量呢，不参与运算吗？
实际上MATLAB是先算完左侧矩阵，然后再回去算右侧矩阵，这里我权当自己是MATLAB，后面再加入$b$的运算，这里第一步消元的位置是$(2,1)$，即row 2,column 1，我们用$E_21$表示，下一步需要完成row 3,column 1的位置，但是此位置已经为0，操作是行1乘以0减去行3，所以无需操作，但是如果MATLAB，这一步是不会省下的，还会做运算。

下面是主元2，他在哪？在$(2,2)$的位置，下面我们想消去$(3,2)$位置，行3减去(行2乘以2)，得到
$\begin{bmatrix}1&2&1\\0&2&-2\\0&0&5\end{bmatrix}$，主元3是$(3,3)$位置，，通过消元我们得到的这个矩阵记做$U$，表示上三角矩阵，从矩阵$A$消元得到矩阵$U$,是科学计算中最普遍的运算，由于普遍，所以人们开始考虑更快的算法，我们这里用的是最直观的方法，找到三个主元，(主元不能为0)。

下面我们讨论下消元失效的情况，之后再回来有效的讨论

失效，指的是不能得到3个主元，如果在进行消元时，0占据了主元的位置，我们就需要交换行，然后再找出剩下主元。
例如在上述的例子中，如果$a_{22}$的值最开始为6，那么当前$a_{22}$的值就变为0了，而0不能作为主元，但是我可以进行行交换，只要$a_{22}$下存在非0元素，问题就解决了，本例中下面是$a_{32}$是4，所以可以进行行交换后找到3个主元。


那么把最初的$a_{33}$的值改为$-4$，进行完第一步得到矩阵$\begin{bmatrix}1&2&1\\0&2&-2\\0&4&-4\end{bmatrix}$
那么就得不到3个主元了，方程无解，此矩阵也"不可逆"，消元彻底失效

回到消元有效的情况：
下一个注意是：回代，什么是回代呢？
我们现在引入右侧向量$b$，MATLAB会怎么做呢？我们照做就行了，在矩阵$A$右侧引入$b$，得到增广矩阵(augmented matrix)
$\begin{bmatrix}1&2&1&2\\3&8&1&12\\0&4&1&2\end{bmatrix}$,augment就是增加的意思，这里增加了新列，在做运算时，我们track on $b$，那么第一步是什么？从方程二减去三倍方程一，得到$\begin{bmatrix}1&2&1&2\\0&2&-2&6\\0&4&1&2\end{bmatrix}$，进行下一步，得到
$\begin{bmatrix}1&2&1&2\\0&2&-2&6\\0&0&5&-10\end{bmatrix}$，那么我们最终得到$U$对应的右侧向量$b$，我们记做$c$。
$c$是$b$的最终结果，就像$U$对$A$一样，这样消元就全部讲完了

那么什么是回代？将消元完后的矩阵的方程组表示出来
$\begin{cases}x&+2y&=2\\2y&-2z&=6\\5z&=-10\end{cases}$，下面如何求解呢？先求哪一个？
一目了然，先求$z$，$z=-2$,然后通过方程2求得到$y=1$,最后通过方程1得到$x=2$,回代是反向求解方程的简单步骤，
因为方程组是三角的，所以反向。
以上是本讲的第一部分

第二部分
引入消元矩阵(elimination matrix)的概念,
如何进行行向量运算，举例：$\begin{bmatrix}1&2&7\end{bmatrix}$
$\begin{bmatrix}X&X&X\\X&X&X\\X&X&X\end{bmatrix}$=$1*row1$+$2*row1$+$7*row1$
结果仍是一行,这里其实是对行进行组合。
这里需要一个矩阵乘以$A$后得到的效果是subtract 3*row from row2,用矩阵表示即：
$\begin{bmatrix} & & \\ &?& \\ & & \end{bmatrix}$$\begin{bmatrix}1&2&1\\3&8&1\\0&4&1\end{bmatrix}$
= $\begin{bmatrix}1&2&1\\0&2&-2\\0&4&1\end{bmatrix}$,

答案是$\begin{bmatrix}1&0&0\\-3&1&0\\0&0&1\end{bmatrix}$，下面需要检验特定的元素，如何检验矩阵乘法中的特定元素？
比如我想检验运算后的$a_{23}$元素，它是这么来的？只需看前面矩阵的行二与后面矩阵的列三即可。
$-3*1+1*1+0*1=-2$.

现在矩阵乘法，我们已经有好几种方法了！可以用行，用列，用其中的每个元素(entry)。
但是这里左侧的矩阵运算起来很简单，我们称它为"初等矩阵"，因为它是使得位置$(2,1)$的元素为0，所以我们标记这个矩阵为
$E_21$，

我们现在是用矩阵语言来进行消元，那么下一步我们需要一个什么矩阵来消$a_32$呢？
$\begin{bmatrix}1&0&0\\0&1&0\\0&-2&1\end{bmatrix}$
$\begin{bmatrix}1&2&1\\0&2&-2\\0&4&1\end{bmatrix}$=
$\begin{bmatrix}1&2&1\\0&2&-2\\0&0&5\end{bmatrix}$
我们将左侧矩阵记为$E_{32}$，

整个过程为:$E_{32}$($E_{21}$$A$)=$U$,
那么什么矩阵能够一步将$A$转变为$U$，答案显而易见，我们可以改变括号顺序，注意矩阵运算时，矩阵的位置是不能改变的。
($E_{32}$$E_{21}$)$A$=$U$。

这里有另一个初等矩阵的例子，它就是交换两行的矩阵，称为"置换矩阵"(permutation matrix)。

这里举一个2*2矩阵的例子，要exchange rows1 and row2， ，$\begin{bmatrix}0&1\\1&0\end{bmatrix}$$\begin{bmatrix}a&b\\c&d\end{bmatrix}$=
$\begin{bmatrix}c&d\\a&b\end{bmatrix}$

其实要找置换矩阵，用单位矩阵(identity matrix)更简单，对单位矩阵进行行变换就得到置换矩阵了。

如果要进行列交换呢。
$\begin{bmatrix}a&b\\c&d\end{bmatrix}$$\begin{bmatrix}0&1\\1&0\end{bmatrix}$=
$\begin{bmatrix}b&d\\a&c\end{bmatrix}$

对左边矩阵用乘法，进行行变换，对右边矩阵用乘法，进行列变换。
To do column operations,it multiplies on the right.To do row operations,it multiplies on the left.
矩阵运算中结合律(commutative law)成立，但是交换律不成立。

回到($E_{32}$$E_{21}$)$A$=$U$的例子，我们可以通过($E_{32}$$E_{21}$)后得到的矩阵，使得我们可以一步到位的将$A$转换额为$U$
，但是这里还有更好的办法，这个方法不是做$A$to$U$，而是做$U$to$A$，就是进行逆运算。
最后一点内容讲"Invesses"，这里举一个例子就可以了
$\begin{bmatrix}1&0&0\\-3&1&0\\0&0&1\end{bmatrix}$我希望找到矩阵来做回这一步，消除这个矩阵的效果，也就是与整个矩阵做乘法得到一个单位矩阵。
如果这个步骤是行二减去3倍行一，那么逆操作就是行二加上三倍行一：
$\begin{bmatrix}1&0&0\\3&1&0\\0&0&1\end{bmatrix}$
$\begin{bmatrix}1&0&0\\-3&1&0\\0&0&1\end{bmatrix}$=
$\begin{bmatrix}1&0&0\\0&1&0\\0&0&1\end{bmatrix}$
我们记做$E^{-1}$$E$=$I$
